```python
affiliate = ""
num_ticket = 0
num_popcorn = 0
num_soda = 0 
total_ticket = 0
total_soda = 0
total_popcorn = 0
total = 0

def cost_tiket(price = 10000):
  return float(price)

def cost_popcorn(price = 5000):
  return float(price)

def cost_soda(price = 2000):
  return float(price)

def validar_codigo(code):
  message = ''
  if len(code) == 10 and code[3] == '#' and code[0] == code[9]:

    if code.count('*') == 1:
      message = 'Platino'

    elif code.count('*') == 2:
      message = 'Dorado'

    elif code.count('*') == 3:
      message = 'Diamante'

  else:
      message = "Codigo incorrecto."   

  return message

def golden(num_ticket, num_popcorn, num_soda):
  flat_ticket = 0
  flat_popcorn = 0
  flat_soda = 0
  if num_ticket < 5:
      flat_ticket = num_ticket
  else: 
    flat_ticket = 4
  if num_ticket < 5:
      flat_popcorn = num_popcorn
  else: 
    flat_popcorn = 4
  if num_ticket < 5:
      flat_soda = num_soda
  else: 
    flat_soda = 4
  
  total_ticket = flat_ticket * cost_tiket()
  total_ticket = total_ticket - (total_ticket * 0.2)
  if num_ticket > 4:
    total_ticket = total_ticket + ((num_ticket - 4) * cost_tiket())

  total_popcorn = flat_popcorn * cost_popcorn()
  total_popcorn = total_popcorn - (total_popcorn * 0.3)
  if num_popcorn > 4:
    total_popcorn = total_popcorn + ((num_popcorn - 4) * cost_popcorn())

  total_soda = flat_soda * cost_soda()
  total_soda = total_soda - (total_soda * 0.4)
  if num_soda > 4:
    total_soda = total_soda + ((num_soda - 4) * cost_soda())

  total = total_soda + total_ticket + total_popcorn

  print(f'El precio total de las boletas es {total_ticket}')
  print(f'El precio total de las crispetas es {total_popcorn}')
  print(f'El precio total de las gaseosas es {total_soda}')
  print(f'El precio total es {total}')
  
def diamond(num_ticket, num_popcorn, num_soda):
  flat_ticket = 0
  flat_popcorn = 0
  flat_soda = 0
  if num_ticket < 6:
      flat_ticket = num_ticket
  else: 
    flat_ticket = 5
  if num_ticket < 6:
      flat_popcorn = num_popcorn
  else: 
    flat_popcorn = 5
  if num_ticket < 6:
      flat_soda = num_soda
  else: 
    flat_soda = 5
  
  total_ticket = flat_ticket * cost_tiket()
  total_ticket = total_ticket - (total_ticket * 0.5)
  if num_ticket > 5:
    total_ticket = total_ticket + ((num_ticket - 5) * cost_tiket())

  total_popcorn = flat_popcorn * cost_popcorn()
  total_popcorn = total_popcorn - (total_popcorn * 0.4)
  if num_popcorn > 5:
    total_popcorn = total_popcorn + ((num_popcorn - 5) * cost_popcorn())

  total_soda = flat_soda * cost_soda()
  total_soda = total_soda - (total_soda * 0.5)
  if num_soda > 5:
    total_soda = total_soda + ((num_soda - 5) * cost_soda())

  total = total_soda + total_ticket + total_popcorn

  print(f'El precio total de las boletas es {total_ticket}')
  print(f'El precio total de las crispetas es {total_popcorn}')
  print(f'El precio total de las gaseosas es {total_soda}')
  print(f'El precio total es {total}')

def platinum(num_ticket, num_popcorn, num_soda):
  flat_ticket = 0
  flat_popcorn = 0
  flat_soda = 0
  if num_ticket < 4:
      flat_ticket = num_ticket
  else: 
    flat_ticket = 3
  if num_ticket < 4:
      flat_popcorn = num_popcorn
  else: 
    flat_popcorn = 3
  if num_ticket < 4:
      flat_soda = num_soda
  else: 
    flat_soda = 3
  
  total_ticket = flat_ticket * cost_tiket()
  total_ticket = total_ticket - (total_ticket * 0.05)
  if num_ticket > 3:
    total_ticket = total_ticket + ((num_ticket - 3) * cost_tiket())

  total_popcorn = flat_popcorn * cost_popcorn()
  total_popcorn = total_popcorn - (total_popcorn * 0.1)
  if num_popcorn > 3:
    total_popcorn = total_popcorn + ((num_popcorn - 3) * cost_popcorn())

  total_soda = flat_soda * cost_soda()
  total_soda = total_soda - (total_soda * 0.1)
  if num_soda > 3:
    total_soda = total_soda + ((num_soda - 3) * cost_soda())

  total = total_soda + total_ticket + total_popcorn

  print(f'El precio total de las boletas es {total_ticket}')
  print(f'El precio total de las crispetas es {total_popcorn}')
  print(f'El precio total de las gaseosas es {total_soda}')
  print(f'El precio total es {total}')

affiliate = input('Usted es afiliado? ').upper()

if affiliate == 'SI':
  code_card = input('Por favor ingrese el código de la tarjeta ')
  message = validar_codigo(code_card)
  if message != 'Codigo incorrecto.':
    num_ticket = int(input('Ingrese la cantidad de boletas a comprar '))
    num_popcorn = int(input('ingrese cantidad de crispetas '))
    num_soda = int(input('ingrese cantidad de gaseosas '))

    if message == 'Platino':
      platinum(num_ticket, num_popcorn, num_soda)
    elif message == 'Dorado':
      golden(num_ticket, num_popcorn, num_soda) 
    elif message == 'Diamante':
      diamond(num_ticket, num_popcorn, num_soda)
  else:
    print(message)

elif affiliate == 'NO':
  num_ticket = input('Ingrese la cantidad de boletas a comprar ')
  num_popcorn = input('ingrese cantidad de crispetas ')
  num_soda = input('ingrese cantidad de gaseosas ')

  try:
    num_ticket = int(num_ticket)
    num_popcorn = int(num_popcorn)
    num_soda = int(num_soda)

    total_soda = num_ticket * costTiket();
    total_popcorn = num_popcorn * cost_popcorn()
    total_ticket = num_ticket * cost_tiket()
    total = total_soda + total_ticket + total_popcorn
    
    print(f'El precio total de las boletas es {total_ticket}')
    print(f'El precio total de las crispetas es {total_popcorn}')
    print(f'El precio total de las gaseosas es {total_soda}')
    print(f'El precio total es {total}')
  except:
    print('Uno de los valores ingresados no es un entero')
  #print(f"No afiliado {affiliate}")



```
